```python
#Q2 Extracting Tesla Revenue Data Using Webscraping
import requests
from bs4 import BeautifulSoup
import pandas as pd

def get_tesla_revenue():
    url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"
    headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"}
    
    response = requests.get(url, headers=headers)
    if response.status_code != 200:
        print("Failed to retrieve data")
        return None
    
    soup = BeautifulSoup(response.text, 'html.parser')
    tables = soup.find_all("table")
    
    if len(tables) < 2:
        print("Revenue table not found")
        return None
    
    revenue_table = tables[1] 
    rows = revenue_table.find_all("tr")
    
    data = []
    for row in rows[1:]: 
        cols = row.find_all("td")
        if len(cols) < 2:
            continue
        date = cols[0].text.strip()
        revenue = cols[1].text.strip()
        data.append([date, revenue])
    
    df = pd.DataFrame(data, columns=["Date", "Revenue"])
    return df

if __name__ == "__main__":
    df = get_tesla_revenue()
    if df is not None:
        print(df.tail())
```

```python
#Q5 Tesla Stock and Revenue Dashboard
import yfinance as yf
import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates

def get_stock_data(ticker, start="2000-01-01", end="2025-12-31", interval="1mo"):
    stock = yf.Ticker(ticker)
    df = stock.history(start=start, end=end, interval=interval)
    df.reset_index(inplace=True)
    df["Date"] = pd.to_datetime(df["Date"])
    
    return df

def get_revenue_data(ticker):
    url = f"https://www.macrotrends.net/stocks/charts/{ticker}/tesla/revenue"
    headers = {"User-Agent": "Mozilla/5.0"}
    response = requests.get(url, headers=headers)
    if response.status_code != 200:
        return None
    soup = BeautifulSoup(response.text, 'html.parser')
    tables = soup.find_all("table")
    if len(tables) < 2:
        return None
    revenue_table = tables[1]
    rows = revenue_table.find_all("tr")
    data = []
    for row in rows[1:]:
        cols = row.find_all("td")
        if len(cols) < 2:
            continue
        date = cols[0].text.strip()
        revenue = cols[1].text.strip().replace(",", "").replace("$", "")  
        data.append([date, float(revenue) if revenue.replace(".", "").isdigit() else None]) 
    df = pd.DataFrame(data, columns=["Date", "Revenue"])
    
    df["Date"] = pd.to_datetime(df["Date"])
    
    return df.dropna()

tesla_stock_data = get_stock_data("TSLA")
tesla_revenue_data = get_revenue_data("TSLA")

def make_graph(stock_data, revenue_data=None, title="Tesla Stock and Revenue Dashboard"):
    plt.figure(figsize=(12, 6))
    sns.lineplot(x=stock_data["Date"], y=stock_data["Close"], label="TSLA Stock Price", color='blue')
    
    if revenue_data is not None:
        ax2 = plt.gca().twinx()  # Create second y-axis
        sns.lineplot(x=revenue_data["Date"], y=revenue_data["Revenue"], label="TSLA Revenue", color='green', linestyle='dashed', ax=ax2)
        ax2.set_ylabel("Revenue (in millions)")
    
    plt.title(title)
    plt.xlabel("Year")
    plt.ylabel("Stock Price (USD)")
    
    plt.gca().xaxis.set_major_locator(mdates.YearLocator(base=1)) 
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))  
    plt.xticks(rotation=45)
    
    plt.legend(loc='upper left')
    plt.grid()
    plt.show()

if __name__ == "__main__":
    make_graph(tesla_stock_data, tesla_revenue_data, title="Tesla Stock & Revenue (2000 - 2025)")
```
